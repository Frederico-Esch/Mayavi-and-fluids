```python
def delF(t, pos):
    x, y = pos
    mi = 0.5
    g = 10
    l = 5
    r = np.sqrt(x**2 + y**2)
    u = y * np.sin(r)/(r+0.001) - x*r/10
    v = -x * np.sin(r)/(r+0.001) - y*r/10
    #u = y
    #v = -mi*y - g/l * np.sin(x)
    return [u,v]
```

```python
X, Y = np.mgrid[-5:5:100j, -5:5:100j]
```

```python
U, V = delF(np.zeros_like(X), [X, Y])
```

```python
plt.figure(figsize=(10, 10))
plt.quiver(X[::6], Y[::6], U[::6], V[::6], minlength=2, minshaft=2, headwidth=10, headlength=6, scale=75)
```

```python
t = np.linspace(0, 100, 1000);
sol = solve_ivp(delF, t_span=(0, max(t)), y0=[0, 1], t_eval=t)
```

```python
sol_X, sol_Y = sol.y[0], sol.y[1]
```
